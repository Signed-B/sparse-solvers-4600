```python
import numpy as np
import scipy as scipy
from scipy.linalg import lu
import matplotlib as plt
import matplotlib.pyplot
from time import process_time
```

We want to evaluate the performance of different algorithms for varying levels of sparsity.
We need to:
    - come up with a method of creating invertible matrices given a % of sparsity
        - actually no, use one from: https://sparse.tamu.edu/
    - evaluate the runtime of 
        - Gaussian elimination
        - LU factorization
        - nested dissection
        - other methods
    - evaluate how different patterns of sparsity affect runtime
        - two examples from the assignment doc
        - lots of examples from https://sparse.tamu.edu/
        

```python
# 2.2.1.1 
n = 100

row = np.ones(n)

# Sparsity structure 1
A = np.eye(n)
A[0,:] = row
A[:,0] = row

plt.pyplot.spy(A)
plt.pyplot.title('Sparsity of A')
plt.pyplot.show()

# Sparsity structure 2
B = np.eye(n)
B[-1,:] = row
B[:,-1] = row

plt.pyplot.spy(B)
plt.pyplot.title('Sparsity of B')
plt.pyplot.show()

print(np.linalg.det(A))
print(np.linalg.det(B))

```

```python
# Now we compute and time the LU-factorization of each
# Matrix A
A_time_start = process_time()
p, l, u = lu(A)
A_time_end = process_time()
print('Time elapsed for A: ', A_time_end - A_time_start)
print('Resulting matrices')
plt.pyplot.spy(p)
plt.pyplot.title('Sparsity of p for A')
plt.pyplot.show()
plt.pyplot.spy(l)
plt.pyplot.title('Sparsity of l for A')
plt.pyplot.show()
plt.pyplot.spy(u)
plt.pyplot.title('Sparsity of u for A')
plt.pyplot.show()


B_time_start = process_time()
p, l, u = lu(B)
B_time_end = process_time()
print('Time elapsed for B:', B_time_end - B_time_start)
plt.pyplot.spy(p)
plt.pyplot.title('Sparsity of p for B')
plt.pyplot.show()
plt.pyplot.spy(l)
plt.pyplot.title('Sparsity of l for B')
plt.pyplot.show()
plt.pyplot.spy(u)
plt.pyplot.title('Sparsity of u for B')
plt.pyplot.show()

```
